### 讀取資料
#### 使用pandas讀取資料

```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定資料讀取路徑
dir_data = '/home/edith/文件/100days-data/' #讀取目錄路徑設定
f_app = os.path.join(dir_data,'application_train.csv') #指定目錄的某csv檔
#print(f_app)
app_train = pd.read_csv(f_app) #用pandas讀取csv檔
app_train.head() # 調查讀取下來的資料前五筆內容是什麼
```

```python
#資料有幾欄幾列 info可以看資料基本資訊
app_train.info()
```

```python
#取得所有欄位的名稱
app_train.columns
```

```python
#資料排序
#以NAME_CONTRACT_TYPE欄位做排序後，取出前6筆資料
app_train.sort_values(by = 'NAME_CONTRACT_TYPE').head(6) 
```

```python
#選取資料中某單一欄位
#取#以NAME_CONTRACT_TYPE欄位資料，後10筆
app_train['NAME_CONTRACT_TYPE'].tail(10)
```

```python
#選取資料中某些欄位（非單一欄位），則用list
#選取SK_ID_CURR,NAME_CONTRACT_TYPE欄位前六筆資料
app_train[['SK_ID_CURR','NAME_CONTRACT_TYPE']].head(6)
```
